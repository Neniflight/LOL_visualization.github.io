```python
import pandas as pd
import numpy as np
```

```python
league_df_2 = pd.read_csv("../data/2023_LoL_esports_match_data_from_OraclesElixir.csv")
league_df_2['league'].unique()
```

```python
major_leagues = league_df_2[league_df_2['league'].isin(['LCK', 'LEC', 'LCS', 'LPL'])][['gameid', 'league', 'year', 'split', 'patch','participantid', 'playername', 'playerid', 'teamname', 'teamid', 'champion', 'ban1', 'ban2', 'ban3', 'ban4', 'ban5', 'result', 'cspm', 'monsterkills', 'earned gpm', 'kills','deaths', 'assists', 'vspm', 'teamkills', 'teamdeaths']]
major_leagues
```

```python
cleaned_df = major_leagues.dropna(subset=['cspm'])
cleaned_df.head(5)
```

```python
cleaned_df = cleaned_df.reset_index(drop=True)
```

```python
#since data inputted sequentially, it is a safe bet to impute the patch with the rows before
def fill_patch(df):
    for i in range(df.shape[0]):
        if np.isnan(df.loc[i, 'patch']):
            df.loc[i, 'patch'] = df.loc[i-1, 'patch']
        else:
            continue
fill_patch(cleaned_df)
```

```python
cleaned_df.isna().sum()
#the playername and playerid missingness are due to how the data is inputted with participant id, so it is ok to be missing
# for the bans, the missing data is so minisucule that it is fine to keep them. 
```

```python
team_df = cleaned_df[(cleaned_df['participantid'] == 100) | (cleaned_df['participantid'] == 200)]
players_df = cleaned_df[~((cleaned_df['participantid'] == 100) | (cleaned_df['participantid'] == 200))]
```

```python
team_df.to_csv("../data/team_data.csv")
players_df.to_csv("../data/player_data.csv")
```
